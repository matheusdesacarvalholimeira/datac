```python
import pandas as pd

file_path = './licenciamento_sanitario.csv'  
data = pd.read_csv(file_path, delimiter=';')

print("Primeiras linhas dos dados:")
print(data.head())

print("\nAnálise de Dados Faltantes:")
missing_data = data.isnull().sum()
print(missing_data[missing_data > 0])

date_columns = ["data_conclusao", "data_emissao_licenca", "data_entrada", "data_pagamento", "data_validade_licenca"]
data[date_columns] = data[date_columns].fillna("N/A")

data[['latitude', 'longitude']] = data[['latitude', 'longitude']].fillna('NaN')

print("\nDados após tratamento de dados faltantes:")
print(data[date_columns].head())
print(data[['latitude', 'longitude']].head())


def clean_licenciamento(text):
    if pd.isnull(text):
        return "Inexistente"
    text_lower = text.lower()
    if "vigente" in text_lower:
        return "Vigente"
    elif "expirado" in text_lower:
        return "Expirado"
    elif "solicitado" in text_lower:
        return "Solicitado"
    else:
        return "Inexistente"

licenciamento_columns = ["licenciamento_ambiental", "licenciamento_sanitario", "licenciamento_urbanistico"]
for col in licenciamento_columns:
    data[col + "_status"] = data[col].apply(clean_licenciamento)

print("\nDados após padronização dos campos de licenciamento:")
print(data[[col + "_status" for col in licenciamento_columns]].head())

processed_file_path = './processed_licenciamento_sanitario.csv' 
data.to_csv(processed_file_path, index=False, sep=';')
print("\nDados processados foram salvos em:", processed_file_path)
```
